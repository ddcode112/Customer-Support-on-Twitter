### Motivation
Social media platforms have become useful tools for modern businesses to interact with customers. These can be used to get immediate product feedback, build customer loyalty, attract new customers, and attend to customer needs and problems. Twitter, a fast-paced social platform, thus comes into a great place for people to seek out customer support, forming totally different practices and behaviors than before. We want to explore the Customer Support on Twitter (Kaggle) dataset, which has natural conversations to understand modern customer support. By analyzing the data, we expect to see what good practice looks like and provide insights on enhancing customer experience.

### Related Work

Having been inspired by the Professor's lesson from his lectures that, as an Information Systems student, the most critical factor for companies is how one can add value. That triggered our search for understanding the customer experience and how it can be improved.
Given our team members' backgrounds in natural language processing and data science, we started looking for datasets in social information-sharing contexts. As the dataset (Customer Support on Twitter (Kaggle)) has data in the social information sharing context, is more natural in terms of text used, and has a bounded set of subjects handled by each company, it will be helpful to derive new insights. 

### Data

Source: [Customer Support on Twitter (Kaggle)](https://www.kaggle.com/datasets/thoughtvector/customer-support-on-twitter?datasetId=4133&sortBy=voteCount) <br><br>
The data is in one table with 7 columns and 2,811,774 rows. It includes tweets across 108 companies and their customers with the author id, created time, and the relations between tweets. Sensitive information in tweet content is pre-excluded. The dataset is relatively clean, but the text field requires further processing, like filtering out stopwords and converting emojis to be ready for analysis. From Kaggle's records, this data is used for sentiment analyses, response time prediction, and chatbot development.

The different columns are described below. Every conversation included has at least one request from a consumer and at least one response from a company. Which user IDs are company user IDs can be calculated using the inbound field.
<br><br>
<b>tweet_id</b>
<br>A unique, anonymized ID for the Tweet. Referenced by response_tweet_id and in_response_to_tweet_id.
<br><br>
<b>author_id</b>
<br>A unique, anonymized user ID. @s in the dataset have been replaced with their associated anonymized user ID.
<br><br>
<b>inbound</b>
<br>Whether the tweet is "inbound" to a company doing customer support on Twitter. This feature is useful when re-organizing data for training conversational models.
<br><br>
<b>created_at</b>
<br>Date and time when the tweet was sent.
<br><br>
<b>text</b>
<br>Tweet content. Sensitive information like phone numbers and email addresses are replaced with mask values.
<br><br>
<b>response_tweet_id</b>
<br>IDs of tweets that are responses to this tweet, comma-separated.
<br><br>
<b>in_response_to_tweet_id</b>
<br>
ID of the tweet this tweet is in response to, if any.

### Questions

1. What makes a specific response satisfactory?<br>
a. Miscellaneous - Language Style
<br>As the data in the dataset contains the responses provided by some of the best companies, these responses can be analyzed to understand some of the common features in a satisfactory response. Language styles, response speed, helpfulness of answers, and other possible factors can all contribute to customer satisfaction. <br><br>
b. What is the relationship between speed of response to tweets and customer satisfaction? <br>
The created-at of the tweets, along with the responses to those tweets, can be used to understand whether there is a relation between the speed of response and customer satisfaction. The response speed is generally assumed to be highly correlated to customer satisfaction, and this hypothesis can be tested by analyzing the dataset. We will first extract words from tweets and calculate their sentiment scores. Second, we'll check if the problem is solved or is retweeted again. If so, how is the sentiment score different from the first tweet against different response speeds?

2. What are the main reasons customers look for Twitter Support? <br>
Are the problems general or specific? <br>
It's often the last step for customers to seek help from Support Team. Before that, sources like user guides or FAQs should have solved a lot of problems for customers. <br>
If conversations have eventually moved out of the Twitter platform, the problem should be considered specific. 
This can be achieved by using frequency, word clouds and topic modelling for the two categories i.e. general problems and specific problems to understand repeating reasons and the precentage of problems for each category.

3. When do Support Teams receive more tweets from customers? <br>
Effective support service can generally build customer trust and loyalty. We'll analyze the number of tweets over month, weekday, and hour to see if there is a specific pattern or link to events. Thus, brands will be able to arrange the support resources wisely and helps customers solve problems in time.

4. Company specific response analysis<br>
Different companies can have different language styles of responsing. This can be analysed using word count, stopword count, response sentiment and finding other words that are specific to responses of different companies.


### Possible Findings and Implications:
1. What makes a specific response satisfactory?<br>
Understanding what components are common among satisfactory tweets can help improve responses and thus increase the percentage of tweets achieving customer satisfaction. Knowing the exact speed and satisfaction relation can give companies a target goal to work on.

2. What are the main reasons customers look for Twitter Support? <br>
Are the problems general or specific? <br>
By analyzing the customer support tweet frequency, word clouds and topic modelling, companies can know what they should improve. 
This will help companies to identify new and repeating problems affecting the customers.

3. When do Support Teams receive more tweets from customers? <br>
If a sudden rise of tweets align with a company event like a new product release, company should consider arrange more support resources and review the problems. Also, there can be some busy days or hours that most customers tend to seek help from support team then, company can rearrange the shifts, working hours based on the results.

4. What are the most common terms in company responses? Is there a correlation between the average customer sentiment and the average company response sentiment?<br>
Understanding the way different companies respond can help understand response patterns for specific companies which might be indicative of the company profile, positioning in the market and can also help in predicting responses to customer problems.
